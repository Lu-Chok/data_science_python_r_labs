```python
features = pd.read_csv("data/UCI_HAR_Dataset/features.txt", sep=' ', names=['id', 'cols'])
subject_train = np.loadtxt("data/UCI_HAR_Dataset/train/subject_train.txt")
subject_test = np.loadtxt("data/UCI_HAR_Dataset/test/subject_test.txt")
x_train = np.loadtxt("data/UCI_HAR_Dataset/train/X_train.txt")
y_train = np.loadtxt("data/UCI_HAR_Dataset/train/Y_train.txt")
x_test = np.loadtxt("data/UCI_HAR_Dataset/test/X_test.txt")
y_test = np.loadtxt("data/UCI_HAR_Dataset/test/Y_test.txt")
```

1. Merges the training and the test sets to create one data set.

```python
# subject
x_train = np.append(x_train, subject_train.reshape(-1,1), axis=1)
# activity
x_train = np.append(x_train, y_train.reshape(-1,1), axis=1)

# subject
x_test = np.append(x_test, subject_test.reshape(-1,1), axis=1)
# activity
x_test = np.append(x_test, y_test.reshape(-1,1), axis=1)

# 1. Merges the training and the test sets to create one data set.
X = pd.DataFrame(np.append(x_train, x_test, axis=0), columns=features.cols.values.tolist()+['subject', 'activity'])
```

2. Extracts only the measurements on the mean and standard deviation for each measurement.

```python
# X_mean_std = X[, features$V2 %in% features[grepl("mean", features$V2) | grepl("std", features$V2), ]$V2]
X_mean_std = X[
    features[features.cols.str.contains('mean') | features.cols.str.contains('std')].cols.values
]
X_mean_std[:4]
```

3. Uses descriptive activity names to name the activities in the data set

```python
activities = pd.read_csv("UCI HAR Dataset/activity_labels.txt",sep=' ', names=['activity', 'activity_type'])
activities
```

```python
X = X.merge(activities, how='inner', on='activity')
```

4. Appropriately labels the data set with descriptive variable names.

```python
X = X.drop(columns=['activity']).rename(columns={'activity_type':'activity'})
```

5. From the data set in step 4, creates a second, independent tidy data set with the average of each variable for each activity and each subject.

```python
from itertools import repeat
```

```python
X.groupby(['activity','subject']).agg(dict(zip(features[features.cols.str.contains('mean')].cols.values, repeat('mean'))))
```
